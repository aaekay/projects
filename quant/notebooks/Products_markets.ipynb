#### Geometric Random walk in stock prices

```python
# A simulation of asset price path
import numpy as np
base_price =100
tosses = np.random.rand(100)
tosses[tosses > 0.5] = 1.01
tosses[tosses < 0.5] = 0.99
all_prices = [100]
for i in tosses:
    all_prices.append(np.round(all_prices[-1]*i,2))

```

```python
import matplotlib.pyplot as plt

plt.plot(all_prices, linestyle="solid")
```

The above plot tells if we toss a coin and based on head and tail, if we increase and descrease the stock price, then this pattern will follow

Discrete compound rate

```python
r = 0.03
t1 = 100

t = np.linspace(1, t1, 10*t1)
amount = (1+r)**t
plt.figure(figsize=(10,6))
plt.plot(t, amount)
```

Continuous compound interest rate

```python
r = 0.03
m = 100000000000
t1 = 100

# for graph lets say
t_c = np.linspace(1, t1, 10*t1)
amount_c = (1 + (r/m))**(m*t)
plt.figure(figsize=(10,6))
plt.plot(t_c, amount_c, linestyle="dotted")
plt.plot(t, amount, linestyle="solid")
```

Now, you can appreciate the continuosly compound interest is   higher as compared to discrete rate, we can show if the value of interest rate changes then the difference is even greater

```python
r = 0.07
m = 100000000000
t1 = 100
p = 1

# for graph lets say
t = np.linspace(1, t1, 10*t1)
amount_c = p*(1 + (r/m))**(m*t)
plt.figure(figsize=(10,6))
plt.plot(t, amount_c, linestyle="dotted")

amount = p*(1+r)**t
plt.plot(t, amount, linestyle="solid")
```

#### Calculating present value of future money

Let's say that i would be recieving 1,000,000 USD after 2 years. Given the interest rate of 5%. What is the present value of that money

```python
future_value = 1000000
r = 0.05
t = 0
T = 2

present_value = future_value*np.exp(-r*(T-t))
print(present_value)
```
